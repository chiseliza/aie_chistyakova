```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline         
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import RocCurveDisplay
```

```python
df = pd.read_csv("S05-hw-dataset.csv")
df.head()
```

- Таргет распределен в соотношении 3:2 (где первое - дефолт по кредиту)

```python
df = df.drop(columns=["client_id"])
```

```python
X = df.drop(columns=["default"])
y = df["default"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

dummy = DummyClassifier(strategy="most_frequent")  
dummy.fit(X_train, y_train)

y_pred_dummy = dummy.predict(X_test)
y_proba_dummy = dummy.predict_proba(X_test)[:, 1]  

acc_dummy = accuracy_score(y_test, y_pred_dummy)
auc_dummy = roc_auc_score(y_test, y_proba_dummy)

results = pd.DataFrame({
    "Model": ["Dummy (most_frequent)"],
    "Accuracy": [acc_dummy],
    "ROC-AUC": [auc_dummy]
})
results.round(4)
```

DummyClassifier(strategy="most_frequent") предсказывает самый частый класс, в нашем случае это 0.

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=1000))
])

param_grid = {
    "logreg__C": [0.01, 0.1, 1.0, 10.0, 100.0]
}

grid_search = GridSearchCV(
    pipe,
    param_grid,
    cv=5,
    scoring="roc_auc",     
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
c = grid_search.best_params_
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

acc_lr = accuracy_score(y_test, y_pred)
auc_lr = roc_auc_score(y_test, y_proba)
print(c)
```

```python
plt.figure(figsize=(6, 5))
RocCurveDisplay.from_estimator(best_model, X_test, y_test)
plt.title("ROC Curve")
plt.savefig("figures/roc_curve.png")
plt.show()
```

```python
results = pd.DataFrame({
    "Model": ["Dummy (most_frequent)", "Logistic Regression"],
    "Accuracy": [acc_dummy, acc_lr],
    "ROC-AUC": [auc_dummy, auc_lr]
})
results.round(4)
```

- Бейзлайн модель показала низкое качество, по сравнению с логистической регрессией. Accuracy 58.96% соответствует доле наиболее часто встречающегося класса. 

- Логистическая регрессия значительно лучше бейзлайн модели. Accuracy вырос на 20%. ROC-AUC вырос на 28%. Модель предсказывет не только самый частый класс.

- При C=100.0 модель показала наилучший результат по ROC-AUC.

- Логистическая регрессия является лучшей моделью для этой задачи, по сравнению с безлайном, обходя его по всем метрикам.
