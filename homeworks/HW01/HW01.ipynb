```python
import numpy as np
```

```python
a = np.array([1, 3, 5, 1, 3]) #создадим три одномерных массива
b = np.arange(1, 6)  
c = np.ones(5)
a, b, c
```

```python
print(a.shape, b.shape, c.shape) #выводим атрибуты массивов (размерность, число осей, тип данных)
print(a.ndim, b.ndim, c.ndim)
print(a.dtype, b.dtype, c.dtype)
```

```python
print(a[::2]) #выводим массив с шагом 2
print(b[:3])  #выводим подмассив, от начала до 3 элемента невключительно 
print(c[-1])  #выводим последний элемент
```

```python
print(a*b) #поэлементное умножение
print(a.sum() + b.sum() + c.sum()) #сумма всех элементов массивов
```

```python
d = np.array([[1, 2, 3], #создадим два двумерных массивов
              [4, 5, 6],
              [7, 8, 9]])

e = np.eye(3)
d, e
```

```python
print(np.transpose(d)) #транспонированная матрица
```

```python
print(np.dot(d, e)) #матричное умножение
```

```python
print(d.min() + e.max()) #сумма миннимального и максимального элементов массивов
```
